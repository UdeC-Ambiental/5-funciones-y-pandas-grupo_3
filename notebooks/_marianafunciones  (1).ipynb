### üîπ Ejemplo de funci√≥n personalizada
Crea una funci√≥n que clasifique el pH:

```python
# -------------------------------
# 1Ô∏è‚É£ Funci√≥n: Promedio
# -------------------------------
def calcular_promedio(df, columna):
    return df[columna].mean()

# -------------------------------
# 2Ô∏è‚É£ Funci√≥n: Rango (m√°x - m√≠n)
# -------------------------------
def calcular_rango(df, columna):
    return df[columna].max() - df[columna].min()

# -------------------------------
# 3Ô∏è‚É£ Clasificaci√≥n del pH (seg√∫n tu notebook)
# -------------------------------
def clasificar_ph(valor):
    if valor < 6.5:
        return "√Åcido"
    elif valor <= 8.5:
        return "Neutro"
    else:
        return "B√°sico"

df["categoria_pH"] = df["pH"].apply(clasificar_ph)

df.head()
```

### üí° Actividad
- Calcula promedios de variables num√©ricas.
- Aplica al menos una funci√≥n con `apply()` o `groupby()`.

```python
import pandas as pd

# Cargar datos
df = pd.read_csv('../data/4_resultados_agua.csv')

# Seleccionar solo columnas num√©ricas
numericas = df.select_dtypes(include=['float64', 'int64'])

# Calcular promedio y rango usando apply()
promedios = numericas.apply(lambda x: x.mean())
rangos = numericas.apply(lambda x: x.max() - x.min())

# Tabla final
tabla_final = pd.DataFrame({
    "Promedio": promedios,
    "Rango": rangos
})

tabla_final
```
